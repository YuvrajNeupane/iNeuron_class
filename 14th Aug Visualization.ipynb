```python
x = np.linspace(-1,6, 30)
x
```

```python
y=np.linspace(-1,6,30)
y
```

```python
z = x + y
z
```

```python
def sin_fun(x,y):
    return np.sin(np.sqrt(x**2 + y**2))
```

```python
z = sin_fun(x,y)
```

```python
plt.figure(figsize=(20,10))
ax = plt.axes(projection = '3d')
ax.plot3D(x,y,z)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```
