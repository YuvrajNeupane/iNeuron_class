# marge & join operation

```python
data4 = {
    'emp_id':[101,102,103,104],
    'salary':[234,345,200,400],
    'pd':[23,34,56,78]
}
data5 = {
    'emp_id':[101,102,103,105],
    'mobile':[656562,656563232,656562626,65951121],
    'house_num':[32,6548,965,666]
}
df4 = pd.DataFrame(data4)
df5 = pd.DataFrame(data5)
```

```python
df45 = pd.merge(df4,df5) #marge operation onlu margebaes on common columns name data only
df45
```

```python
df46 = pd.merge(df5,df4) # Right and left swap
df46
```

```python
df46 = pd.merge(df5,df4, how = 'right')
df46
```

```python
df_sale = pd.read_csv('D:\s\iNuron\data\sales_data_final.csv')
```

```python
df_sale.head(2)
```

```python
df_sale[''] = df_sale
```

```python
def profit_flag(a):
    if a > 0:
        return 'positive'
    else:
        return 'Negetive'
```

```python
profit_flag(-45)
```

```python
profit_flag(45)
```

```python
df_sale['flag_profit'] = df_sale['profit'].apply(profit_flag)
```

```python
df_sale.head()
```

```python
df_sale['len_cus_name'] = df_sale['customer_name'].apply(len)
```

```python
df_sale.head()
```

```python
def quntity_flag(a):
    if a < 10:
        return 'low'
    elif a > 10 and a <20:
        return 'medium'
    else:
        return'high'
```

```python
df_sale['flag_quntity'] = df_sale['quantity'].apply(quntity_flag)
```

```python
df_sale.head(2)
```

```python
df_sale['square'] = df_sale['quantity'].apply(lambda a: a**2)
```
